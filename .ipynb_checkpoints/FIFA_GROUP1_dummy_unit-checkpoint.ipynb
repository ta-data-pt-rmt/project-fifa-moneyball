```python
df_filter['foot_binary'] = pd.get_dummies(df_filter['foot'],drop_first=True)
df_filter = df_filter.drop(columns=["foot"])
df_filter.head()
# Left foot = 0 & Right foot = 1
```

```python
def units(col):
    if "M" in col:
        return "M"
    elif "K" in col:
        return "K"
    else:
        return "No_value"
```

```python
df_filter["Dummy_units"] = df_filter["Value"].apply(units)
df_filter.head()
```
